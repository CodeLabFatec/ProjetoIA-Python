```python
from ultralytics import YOLO

videoInstance = YOLO("best.pt")
```

```python
videos = ["../videos/video.mp4", "../videos/video-lado.mp4", "../videos/video-novo.mp4", "../videos/video-sementrar.mp4"]
acuracias = list()

for video in videos:
    results = videoInstance.predict(source= video, stream= True)

    total_confidence = 0
    total_detections = 0

    for result in results:
        for box in result.boxes:
            total_confidence += box.conf
            total_detections += 1

    if total_detections > 0:
        average = total_confidence / total_detections
        acuracias.append(average)
```

```python
for i in range(0,4):
    acuracia = acuracias[i]
    acuracia_formatada = acuracia[0].item() * 100
    video = videos[i]
    print(f'Acurácia do vídeo {video} é de {acuracia_formatada:.2f}%')
```
